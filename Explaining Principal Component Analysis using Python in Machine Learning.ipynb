```python
#Dropping the correlated features to avoid multicollinearity issue.
data = data.drop(["age","dis","rad","tax"], 1)
data.shape
```

```python
#creating feature and target variable
X = data.drop("medv", 1).values
y= data["medv"].values
```

```python
#nstantiate StandardScaler and use .fit() and .transform() methods. We must make sure that we standardize the data by transforming it onto a unit scale (mean=0 and variance=1).
scaler = StandardScaler()
X_std=scaler.fit_transform(X)
print(X_std)
```

```python
# We can alos write a function to find a Covariance
def cov(x, y):
    xbar, ybar = x.mean(), y.mean()
    return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

# Defining a function to create a Covariance matrix
def cov_mat(X):
    return np.array([[cov(X[0], X[0]), cov(X[0], X[1])], \
                     [cov(X[1], X[0]), cov(X[1], X[1])]])


cov_mat(X.T) 
## Calculate covariance matrix using NumPy's .cov()
(np.cov(X.T))
```

```python
#For creating the covariance matrix manually
mean_X = np.mean(X_std, axis=0)
cov_mat = (X_std - mean_X).T.dot((X_std - mean_X)) / (X_std.shape[0]-1)
print('Covariance matrix is {}'.format(cov_mat))
```

```python
#Creating Covariance Matrix using NumPy's .cov()
print('covariance matrix is {}' .format(np.cov(X_std.T)))
```

```python
#Finding Eigen Values and Eigen vectors
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('Eigenvectors: {}' .format(eig_vecs))
print('\nEigenvalues: {}' .format(eig_vals))
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# print the list as of eigenvalues sorted by decreasing order
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
#finding the percentage of variance by each princiapl components
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
```

```python
#Plotting the Principal components in respect of descending order of variance
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(9), var_exp, alpha=0.5, align='center', label='explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```
